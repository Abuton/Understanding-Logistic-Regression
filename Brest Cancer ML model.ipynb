```python
cols_2_drop = ['id', 'Unnamed: 32']
# drop useless columns from the data
df = df.drop(columns=cols_2_drop, axis=1)
for col in cols_2_drop:
    print(f"{col} has been removed from the data")
```

Since the data has been cleaned, modelling is the next thing..

```python
df.info()
```

Get features and Target

```python
features = df.iloc[:, 1:]
target = df.iloc[:, 0]

print(f"There are {len(features.columns)} independent features (predictors) to be used to predict cancer diagnosis")
```

```python
# encode the target 'diagnosis'
from sklearn.preprocessing import LabelEncoder
target_encoder = LabelEncoder()

target_encoded = target_encoder.fit_transform(target.values)
print('target encoded')
```

Splitting the data into train set and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target_encoded, test_size=.2, random_state=2011)

print(X_test.shape, X_train.shape, y_train.shape, y_test.shape)
```

### Building the ML model

**Logistic Regression**

```python
# using logistic Regression
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

logistic_reg_model = LogisticRegression(max_iter=200)

logistic_reg_model.fit(X_train, y_train)
```

```python
# predict on test set

prediction = logistic_reg_model.predict(X_test)
```

```python
# evaluate model
from sklearn import metrics
acc_score = metrics.accuracy_score(y_test, prediction)
f1_score = metrics.f1_score(y_test, prediction)
classification_report = metrics.classification_report(y_test, prediction)
confusion_matrix = metrics.confusion_matrix(y_test, prediction)
```

```python
# see results

print(f'Accuracy Score {acc_score}')
print(f'F1 Score {f1_score}')
print(f'Classification Report\n {classification_report}')
print(f'Confusion MatriX\n {confusion_matrix}')
```

The question now is what is feature importance??

Feature Importance refers to a class of techniques for assigning scores to input features to a predictive model that indicates the relative importance of each feature when making a prediction. Feature importance can be calculated for regression and classification problems.

These scores are useful because theses scores provides us with
1. Better Understanding the data
2. Better Understanding of the model
3. Reducing number of input features (complexity)

```python
# calculating feature importance

importance = logistic_reg_model.coef_[0]
feature_names = features.columns.values
fea_imp = pd.DataFrame(zip(feature_names, importance), columns=['Features', 'Importance Score'])
```

```python
fea_imp = fea_imp.sort_values(by='Importance Score', ascending=False)
```

```python
# function to select most important features
def get_most_important_features(fea_imp):
    """this function select the most important features based on
       feature importance. returns the top 10 important features
       
       input args: fea_imp -- a dataframe of all the features
       return: 10 features as a list
       """
    features_names = fea_imp['Features'][:10]
#     print(f"The feature names are: {features_names}")
    return list(features_names)

def get_features(df, feature_names, target_name):
    """get features and returns the features data and target data"""
    features = df[feature_names]
    target = df[target_name]
    target = target_encoder.fit_transform(target)
    return features, target

def split_data(features, target):
    """ split the data
    """
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=.2, random_state=23)
    return X_train, X_test, y_train, y_test

# function to train the model
def train_model(train, target):
    """ this function builds the model 
    inputs: train : train features
            target : target features
    returns: trained model
    """
    log_model = LogisticRegression(max_iter=200)
    log_model.fit(train, target)
    return log_model

# function to test model
def test_model(model, test_data):
    """this function test the model"""
    prediction = model.predict(test_data)
    return prediction

# function to evaluate model performance
def eval_model(actual, prediction):
    """evaluate the model"""
    acc_score = round(metrics.accuracy_score(actual, prediction), 2)
    f1_score = metrics.f1_score(actual, prediction), 2)
    classification_report = metrics.classification_report(actual, prediction)
    confusion_matrix = metrics.confusion_matrix(actual, prediction)
    
    print(f'Accuracy Score {acc_score}')
    print(f'F1 Score {f1_score}')
    print(f'Classification Report\n {classification_report}')
    print(f'Confusion MatriX\n {confusion_matrix}')
```

```python
# %%timeit
# using the OOP pipeline
feature_names=get_most_important_features(fea_imp)
# get the data
features, target = get_features(df, feature_names, 'diagnosis')
# split data
X_train, X_test, y_train, y_test = split_data(features, target)
# train model
model = train_model(X_train, y_train)
# test model
pred = test_model(model, X_test)
# evaluate model
eval_model(y_test, pred)
```
